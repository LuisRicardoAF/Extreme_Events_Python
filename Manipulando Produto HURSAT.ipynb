### Produtos Hursat

- Element Name: IRWIN
Definition: Infrared window channel brightness temperature
observation. 
The data is provided in scaled
termperature. 
The original B1 data were calibrated
using the ISCCP absolute calibration (version was
that as of May 2006). 
The data were further
calibrated using coincident HIRS observations and
provided corrections to the ISCCP calibration. The
application of the correction is described below.

Depending on the availability of channels the following channels
are optional for all data sets.

- Element Name: VSCHN (optional)
Definition: Visible channel observations.

- Element Name: IRWVP (optional)
Definition: Infrared window channel observations.

- Element Name: IRSPL (optional)
Definition: Split window channel observations.

- Element Name: IRNIR (optional)
Definition: Near-infrared channel observations.

```python
##Plotando os diferentes produtos

#def produtosHursat()
list_IRWIN = []
data_1 = dataset_hurst.variables['VSCHN'][:]
data_2 = dataset_hurst.variables['IRWVP'][:]
data_3 = dataset_hurst.variables['IRSPL'][:]
data_4 = dataset_hurst.variables['IRNIR'][:]


list_IRWIN.append(data_1)
list_IRWIN.append(data_2)
list_IRWIN.append(data_3)
list_IRWIN.append(data_4)

def plotHursatProduts(lista_Prods):
    vetorIMG = lista_Prods[0].reshape(301,301)
    plt.figure(figsize=(20,30))
    plt.subplot(141)
    plt.imshow(lista_Prods[0].reshape(301,301),cmap='gray')
    plt.title('VSCHN')
    plt.subplot(142)
    plt.imshow(lista_Prods[1].reshape(301,301),cmap='gray')
    plt.title('IRWVP')
    plt.subplot(143)
    plt.imshow(lista_Prods[2].reshape(301,301),cmap='gray')
    plt.title('IRSPL') 
    plt.subplot(144)
    plt.imshow(lista_Prods[3].reshape(301,301),cmap='gray')
    plt.title('IRNIR')
    plt.show()
    
plotHursatProduts(list_IRWIN)    
```
